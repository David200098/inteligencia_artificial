Una serie de tiempo es una serie de puntos de datos ordenados en el tiempo. En una serie de tiempo, el tiempo es generalmente la variable independiente y el objetivo es usualmente hacer un pronostico para el futuro

Autocorrelacion
Es la similitud entre observaciones como una funcion del retraso de tiempo entre ellos.


Estacionalidad (Seasonality)
Refiere a las fluctuaciones periodicas.

Estacionariedad (Stationarity)
Una serie de tiempo se dice que es estacionaria si sus propiedades estadisiticas no cambian con el tiempo.

### Verificación de estacionariedad y descomposición STL de la serie

```python
plt.figure(figsize=[15,7])
sm.tsa.seasonal_decompose(df_month.Weighted_Price).plot()
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.Weighted_Price)[1])
plt.show()
```

```python
df_month['Weighted_Price_box'], lmbda = stats.boxcox(df_month.Weighted_Price)
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.Weighted_Price)[1])
```

```python
df_month['prices_box_diff'] = df_month.Weighted_Price_box - df_month.Weighted_Price_box.shift(12)
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.prices_box_diff[12:])[1])
```

```python
# Regular differentiation
df_month['prices_box_diff2'] = df_month.prices_box_diff - df_month.prices_box_diff.shift(1)
plt.figure(figsize=(15,7))

# STL-decomposition
sm.tsa.seasonal_decompose(df_month.prices_box_diff2[13:]).plot()   
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.prices_box_diff2[13:])[1])

plt.show()
```

```python
# Initial approximation of parameters using Autocorrelation and Partial Autocorrelation Plots
plt.figure(figsize=(15,7))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(df_month.prices_box_diff2[13:].values.squeeze(), lags=48, ax=ax)
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(df_month.prices_box_diff2[13:].values.squeeze(), lags=45, ax=ax)
plt.tight_layout()
plt.show()
```

```python
# Initial approximation of parameters
Qs = range(0, 2)
qs = range(0, 3)
Ps = range(0, 3)
ps = range(0, 3)
D=1
d=1
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)

# Model Selection
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')
for param in parameters_list:
    try:
        model=sm.tsa.statespace.SARIMAX(df_month.Weighted_Price_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
```

```python
# Best Models
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
print(best_model.summary())
```

### P4.- ¿Qué es ADF_Test, ACF, PACF, ARIMA y SARIMA? ¿Para qué se usan?

ADF_Test:

Es una prueba de raíz unitaria para una muestra de una serie de tiempo. Es un número negativo. Cuanto más negativo es, más fuerte es el rechazo de la hipótesis nula de que existe una raíz unitaria para un cierto nivel de confianza.

ACF.

Es una manera visual para mostrar la correlacion serias en datos que cambian con el tiempo (series de tiempo). La correlacion serial es cuando un error en un punto en el tiempo viaja a un punto subsecuente en el tiempo.

PACF.

La funcion de autocorrelacion parcial es una medida de correlacion entre observaciones de una serie de tiempo que estan separadas por k unidades de tiempo.

ARIMA.

El modelo arima permite describir un valor como una funcion lineal de datos anteriores y errores debidos al azar, ademas, puede incluir un componente ciclico o estacional.

SARIMA

Los modelos SARIMA captan el comportamiento puramente estacional de una serie

```python
# Inverse Box-Cox Transformation Function
def invboxcox(y,lmbda):
    if lmbda == 0:
        return(np.exp(y))
    else:
        return(np.exp(np.log(lmbda*y+1)/lmbda))
```
